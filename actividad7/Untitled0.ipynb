```python
import numpy as np
from scipy.fftpack import fft
# Number of samplepoints
N = 100
# sample spacing
T = 1.0 / 10.0
x = np.linspace(0.0, N*T, N)
#eliminando la suma
y = np.sin((2*np.pi*x)/10) + np.sin((8*np.pi*x)/10)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.grid()
plt.show()

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.fftpack 

# sample points
N = 744

# sample spacing
T = 1.0/372.0

# example function
x = np.linspace(0.0, N*T, N)
y = 1.5 * np.sin(5*2*np.pi*x) + 3 * np.sin(6*2*np.pi*x)

df = pd.DataFrame({u'hour': x, u'altura(mm)': y})

sumy = sum(abs(df[u'altura(mm)'])**2)

df['FFT (Y)'] = scipy.fftpack.fft(df[u'altura(mm)'])/(N/2)
df['FFT (freq)'] = scipy.fftpack.fftfreq(N, T)

sumffty = sum(abs(df['FFT (Y)'])**2)

print sumy, sumffty
```
