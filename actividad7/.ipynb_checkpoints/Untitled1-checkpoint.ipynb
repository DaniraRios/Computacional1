```python
import pandas as pd

import numpy as np

import matplotlib as plt
df = pd.read_csv("tiburon2016.csv")
```

```python
from datetime import datetime
df['date']= df.apply(lambda x:datetime.strptime("{0} {1} {2} {3}".format(x[u'anio'],x[u'mes'], x[u'dia'], x[u'hora(utc)']), "%Y %m %d %H"),axis=1)
```

```python
df_short=df.drop(df.columns[[0,1,2,3]],axis=1)
```

```python
df_short['altura(mm)']= df_short['altura(mm)'].astype(float).interpolate(method='spline', order=2)
```

```python
df_short['month']=pd.DatetimeIndex(df_short[u'date']).month
```

```python
df_short['day']=pd.DatetimeIndex(df_short[u'date']).day
```

```python
df_enero=df_short.loc[df_short[u'month'] == 1] 
```

```python
df_dias=df_enero.loc[df_enero[u'day'] <6] 
```

```python
import pandas as pd
time_range = pd.date_range('2016-01-01 00:00:00','2016-01-05 23:00:00', freq='H')
# podemos aun seleccionar un rango de días
 
hours = time_range.hour
days = time_range.day
# Podemos calcular el número de horas transcurridas desde la fecha inicial
time_since_beginning = (days-1)*24+hours
time_since_beginning
```

```python
df_dias['hour']=time_since_beginning
```

```python
df_dias
```

```python
del df_dias[u'month']
```

```python
del df_dias[u'day']
```

```python
del df_dias[u'date']
```

```python
df_dias
```

```python
%matplotlib inline
import numpy as np
from scipy.fftpack import rfft
# Number of samplepoints
N = 119
# sample spacing
T = 1.0 / 119
x = df_dias[u'hour']
#eliminando la suma
y = df_dias[u'altura(mm)']
yf = rfft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.grid()
plt.show()

plt.plot(x,y)
plt.grid()
plt.show()
```
