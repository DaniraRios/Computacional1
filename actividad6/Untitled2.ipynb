```python
import pandas as pd

import numpy as np

import matplotlib as plt

df = pd.read_csv("DICFEB.csv")
```

```python
df.columns
```

```python
df['fecha'] = df[u'Date']+ ' ' + df[u'Time']
```

```python
df['date']=pd.to_datetime(df['fecha'], format='%d/%m/%Y %H:%M')
```

```python
df.set_index(['date'],inplace=True)
```

```python
del df['fecha']
```

```python
del df[u'Date']
```

```python
del df [u'Time']
```

```python
%matplotlib inline
```

```python
df
```

```python
import pandas as pd

import numpy as np

import matplotlib as plt
import matplotlib.pyplot as mplt

ax=df.plot()
ax.set_xlim(pd.Timestamp('2014-01-01 00:00:00'), pd.Timestamp('2014-01-30 00:00:00'))
mplt.ylabel('Nivel (m)')
mplt.xlabel('Fecha')
mplt.title('Nivel del mar; Enero 2014')
```
