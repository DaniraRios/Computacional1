```python
import pandas as pd

import numpy as np

import matplotlib as plt

df = pd.read_csv("OCTDIC2013.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df['fecha'] = df[u'Date']+ ' ' + df[u'Time']
```

```python
df
```

```python
df['date']=pd.to_datetime(df['fecha'], format='%m/%d/%Y %H:%M')
```

```python
df
```

```python
df.set_index(['date'],inplace=True)
```

```python
del df['fecha']
```

```python
del df[u'Date']
```

```python
del df[u'Time']
```

```python
df 
```

```python
%matplotlib inline
```

```python
import pandas as pd

import numpy as np

import matplotlib as plt
import matplotlib.pyplot as mplt

ax=df.plot()
ax.set_xlim(pd.Timestamp('2013-10-13 15:30:00'), pd.Timestamp('2013-11-13 15:30:00'))
mplt.ylabel('Nivel (m)')
mplt.xlabel('Fecha')
mplt.title('Nivel del mar-Sargento; Octubre-Noviembre 2013')
```
